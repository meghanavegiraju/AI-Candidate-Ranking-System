```python
def get_skills(candidate):
    skills = []

    for skill in candidate["skills"]:
        skills.append(
            f"{skill['name']} ({skill['proficiency']})"
        )

    return ", ".join(skills)
```

```python
print(get_skills(candidates[0]))
```

```python
def get_education(candidate):

    education = []

    for edu in candidate["education"]:

        education.append(
            f"{edu['degree']} in {edu['field_of_study']} "
            f"from {edu['institution']} "
            f"({edu['start_year']}-{edu['end_year']})"
        )

    return "\n".join(education)
```

```python
print(get_education(candidates[0]))
```

```python
def get_career(candidate):

    career = []

    for job in candidate["career_history"]:

        career.append(
            f"""
Role: {job['title']}
Company: {job['company']}
Industry: {job['industry']}
Duration: {job['duration_months']} months

Description:
{job['description']}
"""
        )

    return "\n".join(career)
```

```python
print(get_career(candidates[0]))
```

```python
def get_certifications(candidate):

    if len(candidate["certifications"]) == 0:
        return "No Certifications"

    return ", ".join(candidate["certifications"])
```

```python
print(get_certifications(candidates[0]))
```

```python
def get_languages(candidate):

    languages = []

    for lang in candidate["languages"]:

        languages.append(
            f"{lang['language']} ({lang['proficiency']})"
        )

    return ", ".join(languages)
```

```python
print(get_languages(candidates[0]))
```

```python
def get_behavior(candidate):

    signals = candidate["redrob_signals"]

    return f"""
Open To Work: {signals['open_to_work_flag']}
GitHub Activity: {signals['github_activity_score']}
Recruiter Response Rate: {signals['recruiter_response_rate']}
Interview Completion Rate: {signals['interview_completion_rate']}
Profile Completeness: {signals['profile_completeness_score']}
"""
```

```python
print(get_behavior(candidates[0]))
```

```python
SKILL_MAP = {

    "Python": [
        "Python"
    ],

    "LLM": [
        "Fine-tuning LLMs",
        "LoRA",
        "PEFT",
        "Prompt Engineering"
    ],

    "Embeddings": [
        "Sentence Transformers",
        "Embeddings",
        "BGE",
        "E5",
        "OpenAI Embeddings"
    ],

    "Vector Database": [
        "Milvus",
        "FAISS",
        "Pinecone",
        "Qdrant",
        "Weaviate"
    ],

    "Machine Learning": [
        "TensorFlow",
        "PyTorch",
        "Scikit-learn",
        "Machine Learning"
    ],

    "NLP": [
        "NLP",
        "BERT",
        "RoBERTa",
        "T5"
    ],

    "Cloud": [
        "AWS",
        "Azure",
        "GCP"
    ]
}
```

```python
def normalize_skills(candidate):

    normalized = set()

    candidate_skill_names = [
        skill["name"]
        for skill in candidate["skills"]
    ]

    for category, values in SKILL_MAP.items():

        for value in values:

            if value in candidate_skill_names:
                normalized.add(category)

    return normalized
```

```python
normalize_skills(candidates[0])
```

```python
def extract_required_skills(job_text):

    skills = []

    for category, values in SKILL_MAP.items():

        if category.lower() in job_text.lower():
            skills.append(category)

        else:

            for value in values:

                if value.lower() in job_text.lower():
                    skills.append(category)
                    break

    return sorted(list(set(skills)))
```

```python
required_skills = extract_required_skills(job_text)

print(required_skills)
```

```python
def calculate_skill_score_v2(candidate, required_skills):

    candidate_skills = normalize_skills(candidate)

    matched = list(
        set(required_skills).intersection(candidate_skills)
    )

    score = len(matched) / len(required_skills)

    return score, matched
```

```python
score, matched = calculate_skill_score_v2(
    candidates[0],
    required_skills
)

print(score)
print(matched)
```

```python
AI_ROLES = [
    "AI Engineer",
    "ML Engineer",
    "Machine Learning Engineer",
    "Data Scientist",
    "Research Engineer",
    "NLP Engineer",
    "Backend Engineer",
    "Analytics Engineer",
    "Data Engineer"
]
```

```python
AI_TECHNOLOGIES = [
    "Python",
    "Spark",
    "Airflow",
    "PyTorch",
    "TensorFlow",
    "NLP",
    "LLM",
    "Embeddings",
    "LoRA",
    "Milvus",
    "FAISS",
    "Kafka",
    "AWS",
    "GCP",
    "Azure",
    "Vector Database"
]
```

```python
def calculate_experience_score(candidate):

    score = 0

    # Years of experience (maximum 0.4)
    years = candidate["profile"]["years_of_experience"]
    score += min(years / 10, 1) * 0.4

    # Relevant job titles (maximum 0.3)
    for job in candidate["career_history"]:
        if job["title"] in AI_ROLES:
            score += 0.15

    # Relevant technologies in career descriptions (maximum 0.3)
    descriptions = " ".join(
        job["description"] for job in candidate["career_history"]
    ).lower()

    tech_matches = 0

    for tech in AI_TECHNOLOGIES:
        if tech.lower() in descriptions:
            tech_matches += 1

    score += min(tech_matches / 10, 1) * 0.3

    return min(score, 1.0)
```

```python
experience_score = calculate_experience_score(candidates[0])

print("Experience Score:", round(experience_score, 3))
```

```python
def generate_reason(candidate,
                    semantic_score,
                    skill_score,
                    behavior_score,
                    experience_score,
                    matched_skills):

    reasons = []

    # Semantic
    if semantic_score > 0.70:
        reasons.append("Excellent semantic match with the job description.")
    elif semantic_score > 0.50:
        reasons.append("Good semantic match with the job description.")

    # Skills
    if matched_skills:
        reasons.append(
            "Matched skills: " + ", ".join(matched_skills)
        )

    # Experience
    years = candidate["profile"]["years_of_experience"]

    if years >= 5:
        reasons.append(
            f"{years} years of relevant experience."
        )

    # Behavioral
    github = candidate["redrob_signals"]["github_activity_score"]

    if github >= 7:
        reasons.append(
            "Strong GitHub activity."
        )

    if candidate["redrob_signals"]["open_to_work_flag"]:
        reasons.append(
            "Open to work."
        )

    return " ".join(reasons)
```
