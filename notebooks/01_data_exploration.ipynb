# Dataset Summary

| Section | Description | Will be used for Ranking? |
|----------|-------------|---------------------------|
| candidate_id | Unique ID of each candidate | Yes |
| profile | Basic information like name, title, summary, experience | Yes |
| career_history | Previous companies, roles, responsibilities | Yes |
| education | Degree, field of study, institution | Yes |
| skills | Technical skills, proficiency, endorsements | Yes |
| certifications | Professional certifications | Yes |
| languages | Languages known | Optional |
| redrob_signals | Behavioral and recruiter activity metrics | Yes |

## Key Observations

- Every candidate has a unique `candidate_id`.
- The `profile` section contains the candidate's summary and experience.
- `career_history` provides work experience and job descriptions.
- `skills` include technical skills, proficiency level, and endorsements.
- `education` contains degree, institution, and field of study.
- `redrob_signals` include recruiter engagement, GitHub activity, notice period, and open-to-work status.
- These fields will be used to build the candidate ranking system.
