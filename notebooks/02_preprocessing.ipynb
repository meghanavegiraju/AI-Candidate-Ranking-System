```python
import json
import pandas as pd
```

```python
with open("../data/sample_candidates.json", "r", encoding="utf-8") as f:
    candidates = json.load(f)

print("Total Candidates:", len(candidates))
```

```python
def create_candidate_profile(candidate):
    profile = candidate["profile"]

    skills = ", ".join(
        skill["name"] for skill in candidate["skills"]
    )

    career = " ".join(
        job["description"] for job in candidate["career_history"]
    )

    education = " ".join(
        edu["degree"] + " " + edu["field_of_study"]
        for edu in candidate["education"]
    )

    profile_text = f"""
    Headline: {profile['headline']}

    Summary:
    {profile['summary']}

    Experience:
    {profile['years_of_experience']} years

    Current Title:
    {profile['current_title']}

    Skills:
    {skills}

    Career:
    {career}

    Education:
    {education}
    """

    return profile_text
```

```python
text = create_candidate_profile(candidates[0])

print(text)
```

```python
candidate_profiles = []

for candidate in candidates:
    profile = create_candidate_profile(candidate)
    candidate_profiles.append(profile)

print("Profiles Created:", len(candidate_profiles))
```
